```python
import astropy.constants as c
import astropy.units as u
import numpy as np
import math

import matplotlib.pyplot as plt

T=2.76*u.K
def planck_formula(T, freq):
    return (2 * c.h * freq**3 / c.c**2) * 1 / (np.exp(c.h * freq / (c.k_B * T)) - 1)

# logarithmically spaced inputs from 1 to 10^10
x = np.logspace(0, 10, 1000).astype(np.longdouble)*u.GHz
y = x*planck_formula(T, x)

plt.loglog(x,y)
#plt.ylim(1e-40, 1e1)

plt.title=f"Planck spectrum for T={T}"
```
