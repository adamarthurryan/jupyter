```python
%display latex
a=var('a')
b=var('b')
f(x)=(a)/(b^2+x^2)
integrate(f,x)
```

```python
assume(b>0)
integrate(f,x,-oo, oo)
```

```python
xi=var('xi')
g(xi)=integrate(f(a=b/pi),x, -oo, xi)
g
```

```python
plot(g(b=2, a=b/pi), (xi,-200, 200))
```

```python
h(xi)=1/2+arctan(xi/b)/pi
plot(h(b=2), (xi, -3, 3))
```
