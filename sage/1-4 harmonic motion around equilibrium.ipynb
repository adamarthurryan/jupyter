```python
%display latex

a=var('a')
E=var('E')
V(x)=E/a^4*x^4 + 4*a*x^3 - 8*a^2*x^2
F=-diff(V,x,1)
DF=-diff(V,x,2)

plot([V(a=1,E=2),F(a=1,E=2), DF(a=1,E=2)],(-10,10), ymin=-100, ymax=100, title="Potential V, Force F and dF/dt")
```

```python
zeros=solve(F,x)
zeros
```

```python
a=1
E=2
xa=zeros[0](a=a, E=E).rhs()
xb=zeros[1](a=a, E=E).rhs()
x0=xb

approx(x)=-DF(x+x0)*(x)
      #v(a=aa,E=E, x=x+x0),
plot([                      F(a=a,E=E), approx(a=a,E=E)], -.25*a, .25*a, title="Approximate Force for Small Displacements from Equilibrium")
```
