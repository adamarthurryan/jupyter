```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.arange(2,1000) #np.logspace(0, 3, 100) # 1 to 1000, log-spaced
y = x**4-x**3
xlog = np.log(x)
ylog = np.log(y)

plt.subplot(2,2,1)
plt.plot(x, y, 'b-', label='y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('linear Plot')
plt.legend()
plt.grid(True, which="both", ls="-")

plt.subplot(2,2,2)
plt.plot(x, ylog, 'b-', label='y = log(x^2)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis (log scale)')
plt.title('linear-log Plot')
plt.legend()
plt.grid(True, which="both", ls="-")

plt.subplot(2,2,4)
plt.plot(xlog, y, 'b-', label='y = log(x^2)')
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis')
plt.title('log-linear Plot')
plt.legend()
plt.grid(True, which="both", ls="-")

# Create a log-log plot
plt.subplot(2,2,3)
plt.loglog(x, y, 'b-', label='y = x^2')
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis (log scale)')
plt.title('Log-Log Plot')
plt.legend()
plt.grid(True, which="both", ls="-")


# Display the plot
plt.tight_layout()
plt.show()
```
