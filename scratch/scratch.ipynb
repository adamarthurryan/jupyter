```python
import astropy.constants as const
import astropy.units as u
import numpy as np

L = 0.04 * u.L_sun
R = 1.5 * u.R_earth
# L = sigma * R**2 * T**4 
# T = (L /(sigma * R**2)) **(1/4)

T = (L/(const.sigma_sb * R**2)) **(1/4)
print(f"temp {T.to(u.K)}")
```

```python
#a_g = G*M / R**2
#a_c = v**2 / R
#P = 2*pi*R / v

a_g = (const.GM_earth / const.R_earth**2)
a_c = a_g/2
v = (a_c * const.R_earth) ** (1/2)
P = 2 * np.pi *const.R_earth / v
print(v)
print(P.to(u.hour))
```

```python
#rho = 3*M/(4*pi*R^3)

rho_sun = 3*const.M_sun / ( 4*np.pi*const.R_sun**3)
rho_wd = 2e6 * u.g / u.cm**3
ratio = (rho_wd/rho_sun)
print(ratio.to(u.dimensionless_unscaled))
```
